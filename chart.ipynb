```python
merged_data_path = '/Users/jibanchaudhary/Documents/Projects/trading_project/merged_data.csv'
data = pd.read_csv(merged_data_path, index_col='Date', parse_dates= True)
for col in ['LTP','% Change','High','Low','Open','Qty.','Turnover']:
    data[col] = pd.to_numeric(data[col].astype(str).str.replace(',', '').astype(float))
data.to_csv('merged_data.csv')
```

```python
# Scaling the data in the range [0,1]
from sklearn.preprocessing import MinMaxScaler

features = ['LTP', 'High', 'Low', 'Open', 'Qty.', 'Turnover']
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(data[features])
n_steps = 3




```

```python
# appending 3 days data and predicting the fourth day LTP,converting these into 3 dimensional array

def create_sequences(data, n_steps):
    x,y = [],[]
    for i in range(len(data)-n_steps):
        x.append(data[i:i+n_steps])
        y.append(data[i+n_steps,0])
    return np.array(x),np.array(y)

x,y = create_sequences(scaled_data,n_steps)
print('Input shape:', x.shape)
print('Output shape:', y.shape)
for i in range(2):
    print("Example Input x[0]",x[i])
    print('Example Target y[0]',y[i])
```

```python
# Splitting the data into 20% test and 80% train
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
print("Training set",x_train.shape)
print("Testing set",x_test.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam

def create_optimized_LSTM_model(input_shape):
    # model = Sequential()
    # model.add(LSTM(20, return_sequences = True, input_shape=input_shape))
    # model.add(LSTM(40, return_sequences = False))
    # model.add(Dense(1, activation = 'linear'))
    # model.compile(loss = 'mse', optimizer = 'rmsprop')
    # return model
    model = Sequential()

    # First LSTM layer with dropout
    model.add(LSTM(units=50, return_sequences=True, input_shape=input_shape))
    model.add(Dropout(0.2))

    # Second LSTM layer without returning sequences
    model.add(LSTM(units=50, return_sequences=False))
    model.add(Dropout(0.2))

    # Dense output layer for regression
    model.add(Dense(units=1, activation='linear'))

    # Compile the model
    model.compile(optimizer=Adam(learning_rate=0.001, clipvalue=1.0), loss='mse')
    return model



# Example usage
input_shape = (x_train.shape[1], x_train.shape[2])  # (timesteps, features)
model = create_optimized_LSTM_model(input_shape)
model.summary()
```

```python
# Train the model
history = model.fit(x_train, y_train, epochs=20, batch_size=16, validation_data=(x_test, y_test))

# Plotting phase
def plot_metric(history,train_metric,val_metric,metric_name):
    plt.plot(history.history[train_metric],label = f'Training {metric_name}')
    plt.plot(history.history[val_metric],label = f'validation{metric_name}')
    plt.xlabel('Epochs')
    plt.ylabel(metric_name)
    plt.legend()
    plt.show()
plot_metric(history,'loss','val_loss','Loss')
plot_metric(history,'accuracy','val_accuracy','Accuracy')

```
